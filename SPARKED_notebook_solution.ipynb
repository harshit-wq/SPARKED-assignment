# Creating our ans dataset named as df

```python
df=pd.DataFrame(columns=['customer_id','product_id'])
```

```python
for i in range(1,4):
    df.loc[len(df.index)]=['C1',f"P{i}"]
```

```python
for i in range(2,8):
    for j in range(1,3):
        df.loc[len(df.index)]=[f"C{i}",f"P{j}"]
```

```python
df
```

```python
customers['loyalty_score']=customers['loyalty_score'].astype('str')
```

```python
df = (df.merge(customers, left_on='customer_id', right_on='customer_id')
          .reindex(columns=['customer_id','loyalty_score','product_id']))
```

```python
df
```

```python
products
```

```python
df = (df.merge(products, left_on='product_id', right_on='product_id')
          .reindex(columns=['customer_id','loyalty_score','product_id','product_category','product_discription']))
```

```python
df=df.drop('product_discription',axis=1)
```

```python
df
```

```python
dict1=df['product_id'].value_counts().to_dict()
```

```python
dict1
```

```python
df['purchase_count_per_product_id']=df.product_id.map(dict1)
```

```python
df=df.sort_values(by='customer_id')
```
